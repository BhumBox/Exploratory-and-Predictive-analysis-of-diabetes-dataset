### Problem 1: Matplotlib

```python
# Pima Indians Diabetes Database
# This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. 
# The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, 
# based on certain diagnostic measurements included in the dataset. 
# Several constraints were placed on the selection of these instances from a larger database. 
# In particular, all patients here are females at least 21 years old of Pima Indian heritage.

import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv('diabetes.csv')
df.head() #prints the first 5 rows
```

```python
# (a) using Matplotlib scatter plot to show Glucose vs Insulin for all the data, add label name to x and y axis


plt.figure(figsize=(10,6))
plt.scatter(df['Insulin'], df['Glucose'])
plt.xlabel('Insulin')
plt.ylabel('Glucose');
```

```python
# (b) using Matplotlib scatter plot to show Glucose vs Insulin for each type of Outcome in different colors
# add legend and label name to x and y axis

df_Out1= df[df['Outcome']==1]
df_Out0=df[df['Outcome']==0]

plt.figure(figsize=(10,6))
plt.scatter(df_Out1['Glucose'],df_Out1['Insulin'],c='r')
plt.scatter(df_Out0['Glucose'],df_Out0['Insulin'],c='y')
plt.legend(['outcome=1','outcome =0'])
plt.xlabel('Glucose')
plt.ylabel('Insulin');
```

```python
# (c) using Matplotlib boxplot to show Glucose for each type of Outcome, add label name to y axis
plt.figure(figsize=(10,6))

Out1= df.loc[df['Outcome']==1,:]
Out0= df.loc[df['Outcome']==0,:]

plt.boxplot([Out1['Glucose'], Out0['Glucose']])
plt.gca().set_xticklabels(['Out1','Out0']);
plt.ylabel('Glucose')
```

```python
# (d) using Matplotlib boxplot to show Insulin for each type of Outcome, add label name to y axis
plt.figure(figsize=(10,6))

Out1= df.loc[df['Outcome']==1,:]
Out0= df.loc[df['Outcome']==0,:]

plt.boxplot([Out1['Insulin'], Out0['Insulin']])
plt.gca().set_xticklabels(['Out1','Out0']);
plt.ylabel('Insulin')
```

### Problem 2: Seaborn

```python
# (a) using seaborn boxplot to show Glucose for each type of Outcome
plt.figure(figsize=(10,6))
sns.boxplot(x='Outcome', y='Glucose', data = df)
```

```python
# (b) using seaborn boxplot to show Insulin for each type of Outcome
plt.figure(figsize=(10,6))
sns.boxplot(x='Outcome', y='Insulin', data = df)
```

```python
# (c) using seaborn pairplot to Plot pairwise relationships for columns ['Glucose','BMI','Age', 'Insulin','SkinThickness'],
# colored by each type of Outcome
plt.figure(figsize=(10,10))
sns.pairplot(df,vars=['Glucose','BMI','Age', 'Insulin','SkinThickness'], hue='Outcome',palette="husl");
```

```python
# (d) using seaborn distplot to show the histogram of all columns 
plt.figure(figsize=(10,10));
plt.subplot(2,2,3)
sns.distplot( df["Glucose"] , color="skyblue", label="Glucose")
sns.distplot( df["BMI"] , color="red", label="BMI")
sns.distplot( df["Insulin"] , color="yellow", label="Insulin")
sns.distplot( df["Pregnancies"] , color="pink", label="Pregnancies")
sns.distplot( df["BloodPressure"] , color="darkblue", label="BloodPressure")
sns.distplot( df["Age"] , color="red", label="green")
sns.distplot( df["SkinThickness"] , color="brown", label="SkinThickness")
sns.distplot( df["DiabetesPedigreeFunction"] , color="orange", label="DiabetesPedigreeFunction")
plt.legend()

#plt.show()

```
